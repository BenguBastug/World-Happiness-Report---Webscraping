```python
# The mean happiness score for each region

df2019_mean = df2019.groupby('Region')[['happiness_score']].mean()
df2019_mean_region = df2019_mean.sort_values(by='happiness_score', ascending=False)
print(df2019_mean_region)
fig, ax = plt.subplots(1,1)
sns.barplot(y = df2019_mean_region.index,
            x = df2019_mean_region.happiness_score,
            orient='h',
            palette="ch:s=.25,rot=-.25",
            ax=ax)
plt.title("Average Happiness Score per Region")
plt.xlabel("Average Happiness Score") 
plt.ylabel("")
plt.xlim(left=4)
plt.grid(axis='x', alpha=.3)
plt.show()
```

```python
# The number of countries in each region that have a happiness score above 6.0

happinest_6 = df2019.loc[(df2019['happiness_score'])>6]
happinest_6 = happinest_6.groupby(['Region'])[['happiness_score']].count()
df2019_happinest_6 = happinest_6.sort_values(by = 'happiness_score', ascending = False)
print(df2019_happinest_6)
fig, ax = plt.subplots(1,1)
sns.barplot(y = df2019_happinest_6.index,
            x = df2019_happinest_6.happiness_score,
            orient='h',
            palette="ch:s=.25,rot=-.25",
            ax=ax)
plt.title("Happiness score above 6.0")
plt.xlabel("Number of countries") 
plt.ylabel("")
plt.xticks([0, 3, 6, 9, 12, 15, 18])
plt.grid(axis='x', alpha=.3)
plt.show()
```

```python
# Which region has the largest range of happiness scores? 

r_max_min = df2019.groupby(['Region'])[['happiness_score']].max()-df2019.groupby(['Region'])[['happiness_score']].min()
df2019_range = r_max_min.sort_values(by = 'happiness_score', ascending = False)
print(df2019_range)
fig, ax = plt.subplots(1,1)
sns.barplot(y = df2019_range.index,
            x = df2019_range.happiness_score,
            orient='h',
            palette="ch:s=.25,rot=-.25",
            ax=ax)
plt.title("Range of Happiness Score per Region")
plt.xlabel("Range of Happiness Score") 
plt.ylabel("")
plt.grid(axis='x', alpha=.3)
plt.show()
```
