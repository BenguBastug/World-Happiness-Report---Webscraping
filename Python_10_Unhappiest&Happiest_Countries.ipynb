```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#lets show the top 10 happiest countries in 2019 :)

df2019=pd.read_csv ('2019.csv')
df2019
sorted_df2019 = df2019.sort_values(by='Score', ascending=False)
top_10_happiest = sorted_df2019.head(10)
top_10_happiest
```

Here I shall display the Bar Plots that have been created for the Streamlit  visualization.

Top 10 Happiest Countries of 2019

```python
df2019=pd.read_csv ('2019.csv')
df2019
sorted_df2019 = df2019.sort_values(by='Score', ascending=False)
top_10_happiest = sorted_df2019.head(10)
top_10_happiest
sns.barplot(x='Country or region', y='Score', data=top_10_happiest)
plt.ylim(6,8)
plt.title('Top 10 Happiest Countries in the World in 2019')
plt.xlabel('')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.show()
```

Top 10 Happiest Countries of 2018

```python
df2018=pd.read_csv ('2018.csv')
df2018
top_10_happiest = sorted_df2018.head(10)
top_10_happiest
sns.barplot(x='Country or region', y='Score', data=top_10_happiest)
plt.ylim(6,8)
plt.title('Top 10 Happiest Countries in the World in 2018')
plt.xlabel('')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.show()
```

Top 10 Unhappiest Countries of 2019

```python
df2019L=pd.read_csv ('2019.csv')
df2019L
sorted_df2019 = df2019L.sort_values(by='Score', ascending=True)
last_10_data = sorted_df2019.head(10)
last_10_data
last_10_data = df2019.tail(10)
sns.barplot(x='Country or region', y='Score', data=last_10_data)
plt.ylim(2,4)
plt.title('Top 10 Unhappiest Countries in the World')
plt.xlabel('')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.show()
```
